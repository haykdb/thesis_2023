```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
meta = pd.read_csv('CRSP_2000.csv', index_col = 0)
meta.head()
```

```python
stocks = set(meta.index)
```

```python
split_val = '2017/01/01'
split_val1 = '2020/01/01'

for i in stocks:
    data = meta[meta.index == i]
    ticker = data.iloc[0, 1]
    data.set_index('date', inplace = True)
    
    start = data.index[0]
    end = data.index[-1]
    
    if start != '2000/01/03' or end != '2022/03/31':
        continue
    else:
        pass
    
    if start < split_val:
        train_data = data.loc[:split_val]
        val_data = data.loc[split_val:split_val1]
        test_data = data.loc[split_val1:]
        
        train_data.to_csv(f'Train_Data/{ticker}.csv', index = True)
        val_data.to_csv(f'Val_Data/{ticker}.csv', index = True)
        test_data.to_csv(f'Test_Data/{ticker}.csv', index = True)
    elif start >= split_val and start < split_val1:    
        val_data = data.loc[:split_val1]
        test_data = data.loc[split_val1:]
        
        val_data.to_csv(f'Val_Data/{ticker}.csv', index = True)
        test_data.to_csv(f'Test_Data/{ticker}.csv', index = True)
    else:
        test_data = data
        test_data.to_csv(f'Test_Data/{ticker}.csv', index = True)
    
    
#     data.to_csv('Data/{}.csv'.format(ticker), index = True)
```
